```python
# 함수 안에서 전역변수의 값을 변경하려면, globla 키워드를 사용해야 한다.
def circle_area():
    global area
    area = 3.14 * radius ** 2
    return area
area = 0
radius = float(input('반지름:'))
print('원의 넓이:',circle_area())
print(area)
```

```python
# 매개변수 개수 가변적일때 사용. 
def average(*args):     # args는 튜플 데이터형
    avrg1 = sum(args) /len(args)
    return avrg1
print(average(1,2,3))
print(average(1,2,3,4,5))
```

```python
def average(*a):
    avrg1 = sum(a) /len(a)
    return avrg1
print(average(1,2,3))
print(average(1,2,3,4,5))
```

```python
def is_prime(number):
    if number < 2:
        return False
    for i in range(2,number):
        if number % i == 0:
            return False
    return True

num = int(input('숫자 : '))
if is_prime(num):
    print('소수입니다.')
else:
    print('소수가 아닙니다')
```

```python
# lambda 함수 : 한줄로 간단하게 만들어주는 함수
def plus(num1,num2):
    result = num1 + num2
    return result
print(plus(10,20))

plus = lambda num1,num2: num1 + num2    # def함수처럼 여러줄을 사용하여 return을 사용하지 않고 간단하게 만듬
print(plus(10,20))
```

```python
# map() 함수
# 두개의 인자를 갖는 함수로 순서가 있는 자료형에 사용
# 첫번째 인자는 함수명, 두번째 인자는 순서가 있는 자료형
```

```python
mylist = [1,2,3,4,5]
plus10 = lambda num: num+10

mylist = list(map(plus10,mylist))   # plus10 은 함수, mylist는 자료형
print(mylist)
```

```python
mylist = [1,2,3,4,5]
mylist = list(map(lambda num:num+10,mylist))
print(mylist)
```

```python
def hello(count):
    if count == 0:  
        return

    print('Hello, world', count)

    count -= 1      # count 를 1 감소시킨 뒤
    hello(count)    # 다시 hello에 넣음
hello(5)
```

```python
def factorial(n):
    if n == 1:       # n이 1일 때
        return 1    # 1을 반환하고 재귀호출을 끝냄
    return n * factorial(n-1)   # n과 factorial 함수에 n-1을 넣어서 반환된 값을 곱함
print(factorial(5))
```

```python
# 편의점 재고관리 시스템 예제
inven = {
    '삼각김밥':10,
    '커피우유':20,
    '콜라':30}
while True:
    print('''
    ===== 재고관리 시스템 =====
    1:재고등록, 2:제품판매, 3:재고확인, 0:프로그램 종료
    ''')

    menu = input('메뉴를 선택하세요 : ')
    if menu == '0':
        print('프로그램을 종료합니다.')
        break
    elif menu == '1':
        print('재고를 등록합니다')
        while True:
            a = input('재고명 : ')
            if a == '':
                break
            elif a in inven.keys():
                b = int(input('갯수 : '))
                inven[a] += b
            else:
                b = int(input('갯수 : '))
                inven[a] = b
    elif menu == '2':
        print('제품을 판매합니다')
        while True:
            c = input('재고명 : ')
            if c == '':
                break
            else:
                d = int(input('갯수 : '))
                if c not in inven.keys():
                    print('없는 제품입니다.')
                elif d > inven[c]:
                    print(f'재고가 부족합니다. (재고량 : {inven[c]}개)')
                else:
                    inven[c] -= d
    elif menu == '3':
        print('재고 확인')
        while True:
            e = input('재고명 : ')
            if e == '':
                break
            elif e in inven.keys():
                print(f'{e} : {inven[e]}개')
            elif e == '전체':
                print(inven)
            else:
                print('없는 제품입니다.')
    else :
        print('메뉴를 다시 입력하세요')
```
